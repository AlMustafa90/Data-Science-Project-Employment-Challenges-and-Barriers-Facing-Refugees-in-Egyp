# Data Science Project: Employment Challenges Facing Refugees in Egypt  

## Introduction  
This project investigates the multifaceted employment challenges facing refugees in Egypt by integrating both quantitative survey data and qualitative text analysis. Data was **collected over a two-month period** through structured questionnaires and open-ended responses, then **prepared and merged initially using Excel** before advanced analysis in Python.  
The study focuses on:  
- **Job unsustainability** among employed refugees  
- **Barriers to entry** faced by unemployed refugees  
- **Narratives and lived experiences** captured through open-ended responses  

---

## Goals & Objectives  
- Identify key factors influencing job suitability, sustainability, and accessibility.  
- Examine barriers such as legal restrictions, discrimination, and limited market opportunities.  
- Apply **Natural Language Processing (NLP)** to extract recurring themes, sentiments, and concerns from refugees’ narratives.  
- Visualize relationships between demographics, qualifications, job types, and employment outcomes.  
- Provide **evidence-based recommendations** for NGOs, policymakers, and employers to enhance refugee employment.  

---

## Dataset  
- **Collection Period:** Two months  
- **Collection Method:** Structured questionnaires + open-ended responses  
- **Preparation Tools:**  
  - *Excel*: for initial cleaning, merging, and tabulation of responses  
  - *Python*: for advanced analytics, NLP, and visualization  

- **Target Groups:**  
  - *Unemployed refugees*: facing challenges in job search and access.  
  - *Employed refugees*: experiencing job–qualification mismatch and unsustainable conditions.  

- **Key Features:**  
  - Demographics: age, gender, education, country of origin  
  - Employment status, job search duration, and suitability assessment  
  - Barriers: legal, language, discrimination, and work conditions  
  - Text data: qualitative feedback and personal experiences  

---

## Analytical Approach  
- **Data Preparation:** Cleaning, transformation, and integration of structured and text data (Excel → Python).  
- **Exploratory Data Analysis (EDA):**  
  - Distribution analysis (age, education, employment status)  
  - Correlation and trend identification (e.g., education vs. job sustainability)  
- **Text Analysis (NLP):**  
  - Sentiment analysis  
  - Topic modeling (e.g., Latent Dirichlet Allocation)  
  - Keyword extraction and word clouds  
- **Comparative Analysis:** Cross-group comparison (employed vs. unemployed).  
- **Predictive Modeling:** Machine Learning to predict **average job search duration** and assess risk factors.  

---

## Key Focus Areas  
### Unemployed Refugees  
- Barriers to labor market integration  
- Average job search duration and influencing factors  
- Legal and social restrictions  
- Psychological and emotional impact  

### Employed Refugees  
- Job–qualification alignment (skills vs. occupation)  
- Employment sustainability and job retention  
- Work conditions and opportunities for growth  
- Sentiment analysis of job satisfaction and experiences  

---

## Expected Outcomes  
- Data-driven insights into employment challenges faced by refugees in Egypt.  
- Thematic analysis of refugee experiences through NLP techniques.  
- Visual dashboards highlighting suitability, accessibility, and sustainability trends.  
- Actionable recommendations for stakeholders to enhance refugee labor market integration.

##### Interaction effect of comparing the 'Academic background' with 'gender' and 'Job suitability' to see the potential impact on job sustainability

```python
import matplotlib.pyplot as plt
import seaborn as sns

filtered_data = Data1[Data1['gender'].isin(['Male', 'Female'])]

plt.figure(figsize=(10, 6))
sns.pointplot(x='Academic background', 
              y='Job suitability', 
              hue='gender', 
              data=filtered_data, 
              dodge=True, 
              markers=['o', 's'], 
              capsize=.1, 
              palette=['#0052B4', '#FFDA44'])  
plt.title('Interaction Effect between Gender and Academic Background on Job Suitability')
plt.xlabel('Academic Background')
plt.ylabel('Average Job Suitability')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency 


factor = 'Academic background'

unique_factors = Data1[factor].unique()

for val in unique_factors:
    subset = Data1[Data1[factor] == val] # 
    contingency_table = pd.crosstab(subset['gender'], subset['Job suitability'])
    

    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    print(f"\nتحليل العامل '{val}':")
    print("قيمة Chi-Square:", chi2)
    print("قيمة P-Value:", p)
    
    if p < 0.05:
        print("نرفض الفرضية الصفرية: هناك علاقة ذات دلالة إحصائية بين Gender و Job Suitability لهذه الفئة.")
    else:
        print("لا نرفض الفرضية الصفرية: لا توجد علاقة ذات دلالة إحصائية بين Gender و Job Suitability لهذه الفئة.")
```

##### analyzing 'Difficulties in getting a job' and 'why jobs are not suitable' 

```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud

# تحميل الموارد الخاصة بـ NLTK
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.isri import ISRIStemmer
```

```python
# إعداد أدوات التنظيف
stop_words = set(stopwords.words('arabic'))
stemmer = ISRIStemmer()

def clean_text(text):
    if pd.isnull(text):
        return ""
    text = ''.join([char for char in text if char.isalpha() or char == ' '])
    text = text.lower()
    text = ' '.join([word for word in text.split() if word not in stop_words]) 
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    return text

# تنظيف الأعمدة
Data1['why it is not suitable'] = Data1['why it is not suitable'].apply(clean_text)
Data1['Difficulties in getting a job'] = Data1['Difficulties in getting a job'].apply(clean_text)
```

```python
# دمج النصوص
text_why_not_suitable = ' '.join(Data1['why it is not suitable'])
text_difficulties = ' '.join(Data1['Difficulties in getting a job'])

# إنشاء السحابة النصية لعمود "why it is not suitable"
wordcloud_why_not = WordCloud(font_path='arial', background_color='white', width=800, height=600).generate(text_why_not_suitable)
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud_why_not, interpolation='bilinear')
plt.axis('off')
plt.title("الكلمات الأكثر تكرارًا في عمود why it is not suitable")
plt.show()

# إنشاء السحابة النصية لعمود "Difficulties in getting a job"
wordcloud_difficulties = WordCloud(font_path='arial', background_color='white', width=800, height=600).generate(text_difficulties)
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud_difficulties, interpolation='bilinear')
plt.axis('off')
plt.title("الكلمات الأكثر تكرارًا في عمود Difficulties in getting a job")
plt.show()
```

```python
# تحويل النصوص إلى مصفوفة TF-IDF
vectorizer_why_not = TfidfVectorizer(max_features=1000)
X_why_not = vectorizer_why_not.fit_transform(Data1['why it is not suitable'])

vectorizer_difficulties = TfidfVectorizer(max_features=1000)
X_difficulties = vectorizer_difficulties.fit_transform(Data1['Difficulties in getting a job'])

# عرض أهم الكلمات
feature_names_why_not = vectorizer_why_not.get_feature_names_out()
tfidf_scores_why_not = X_why_not.sum(axis=0).A1
tfidf_df_why_not = pd.DataFrame({'word': feature_names_why_not, 'score': tfidf_scores_why_not}).sort_values(by='score', ascending=False)

feature_names_difficulties = vectorizer_difficulties.get_feature_names_out()
tfidf_scores_difficulties = X_difficulties.sum(axis=0).A1
tfidf_df_difficulties = pd.DataFrame({'word': feature_names_difficulties, 'score': tfidf_scores_difficulties}).sort_values(by='score', ascending=False)

# عرض النتائج
print("أكثر الكلمات تأثيرًا في why it is not suitable:\n", tfidf_df_why_not.head(10))
print("\nأكثر الكلمات تأثيرًا في Difficulties in getting a job:\n", tfidf_df_difficulties.head(10))
```
